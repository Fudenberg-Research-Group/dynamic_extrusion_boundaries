```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import host_subplot
from polychrom.polymer_analyses import slope_contact_scaling
from scipy.ndimage import gaussian_filter1d

def smooth(x, sigma = 1.75):
        return gaussian_filter1d(x, sigma)
```

```python
### converting from lattice based 
TIME_CONV_COEF = convertor= 21.875 * 60 / 500 #to convert from lattice based to seconds
LATTICE_CONV_COEF = 0.25 #each lattice site is 0.25 kb
```

```python
file = pd.read_csv('../output/scores/simulation/scores/fripscore_sep_100.csv')

file_300 = pd.read_csv('../output/scores/simulation/scores/fripscore_sep_300.csv')
file = file[file[' clife']!=500000]
lifetime = file['lifetime']

unique_life = np.unique(lifetime)
```

### FRiP as a function of ctcf binding time

```python
lifes, vels = [500, 1500] , [1]

for life in lifes:
    for vel in vels:
        new_file = file.query('lifetime == @life and ` velocity` == @vel and ` clife`==10*` cof`')[[' clife',' cof', ' fripscore']]
        new_file_filtered = new_file.drop_duplicates(keep = 'first')
        new_file_filtered = new_file_filtered.sort_values(' clife')
        new_clife = new_file_filtered[' clife']
        new_clife_300 = file_300[' clife']
        frip_score = new_file_filtered[' fripscore']
        frip_score_300 = file_300[' fripscore']
        plt.scatter(new_clife[1:], frip_score[1:], label=r'$\tau_E$ = %s sec'%(np.round(life * TIME_CONV_COEF)))
        plt.plot(new_clife[1:], frip_score[1:], alpha = 0.45)

file_300 = file_300.query('` clife`==10*` cof`')
file_300 = file_300.sort_values(' clife')
plt.scatter(file_300[' clife'][:9],file_300[' fripscore'][:9],color='teal',s=15,label=r'WT, sep = %s kb'%(3000*LATTICE_CONV_COEF))
plt.plot(file_300[' clife'][:9],file_300[' fripscore'][:9],color='teal', alpha = 0.35)
life_w = 500
delta = 350
plt.plot([life_w, life_w, life_w],[0,0.5,0.71],'--', color = 'red',alpha = 0.75, linewidth = 0.75, label = r'$\tau$ = $\tau^W_E$')
plt.plot([life_w-delta, life_w-delta, life_w-delta],[0,0.5,0.71],'--', color = 'blue',
         alpha = 0.75, linewidth = 0.75, label = r'$\tau$$\sim$$\tau^W_E-\delta$')
plt.xlim()
plt.title('occupancy 0.9')
plt.xscale('log')
plt.legend(fontsize = 10.51,loc = (0.001,0.57))
plt.ylabel('FRiP',fontsize = 15)
plt.xlabel(r'$\tau_b$', fontsize = 15)
plt.savefig('../output/figures/frip_for_tau_extruder_params.pdf')
plt.show()
        
```
